### Reading from CSV Files

```python
csv_file_1 = pd.read_csv('species.csv')
csv_file_2 = pd.read_csv('surveys.csv')
```

### Final DataFrame

```python
# Merged the two Dataframes into one DataFrame
Final_df = pd.merge(csv_file_1, csv_file_2, on='species_id')
```

.

## Discovery Analysis

### Taxa discovered per Month over the years

```python
data_all_taxa_month = pd.crosstab(Final_df.month, Final_df.taxa)
```

### Stacked Bar Chart -  Taxa discovered per Month over the years.

```python
# Stacked Bar Chart Legends --
bar_chart_taxa_month = data_all_taxa_month.plot(kind='bar', stacked=True, figsize=(10,10))

bar_chart_taxa_month.set_xlabel("Month", fontsize=15)
bar_chart_taxa_month.set_ylabel("Total Taxa Discovered", fontsize=15)
bar_chart_taxa_month.set_title("All Taxa Discovered per Month", fontsize=18)
```

##### Graph 1.0 : Graph shows the number of taxa discovered per month over the year stacked according to 'Taxa'. This graph shows Rodents are the maximum to be discovered every month. 'July'  being the month when most taxa were discovered


### Count of all the Taxa based on Gender discovered over the years.


```python
bar_plot_all_taxa_gender = pd.crosstab(Final_df.taxa, Final_df.sex).plot(kind='bar')

bar_plot_all_taxa_gender.set_xlabel("Taxa", fontsize=12)
bar_plot_all_taxa_gender.set_ylabel("Total Species", fontsize=12)
bar_plot_all_taxa_gender.set_title("All Taxa Discovered based on Gender over the years", fontsize=15)
```

##### Graph 1.1 : Graph shows the count of taxa discovered over the years stacked according to 'Taxa'. This graph shows only Rodents has been classified as Male and Female. 



#####  So only RODENTS shall be considered for further analysis.

### Count of all the Rodents based on Gender per Month

```python
data_rodent = Final_df[Final_df.taxa == 'Rodent']
```

```python
data_rodent_Sex_Month = pd.crosstab(data_rodent.month, data_rodent.sex)
```


### Stacked Bar Chart - Rodents based on Gender discovered per month.


```python
# Bar Chart Legends --
bar_plot_rodent_gender_month = data_rodent_Sex_Month.plot(kind='bar', stacked=True, figsize=(18,5))

bar_plot_rodent_gender_month.set_xlabel("Month", fontsize=15)
bar_plot_rodent_gender_month.set_ylabel("Total Rodents Discovered", fontsize=15)
bar_plot_rodent_gender_month.set_title("Rodents Discovered based on Gender per month.", fontsize=18)
```

##### Graph 1.2 : Graph shows the count of Rodents discovered per month over the years, based on Gender. This graph shows July Month is the one where maximum Rodents were discovered.

## Weight Analysis


### Wieght of all taxa discovered per month based on Gender over the years 

```python
data_all_taxa_month_gender_weight = Final_df.pivot_table(index =['month'], columns=['taxa','sex'], values=['weight'],)
```

##### Only Rodents has been classified as Male and Female, so Rodents will only be plotted for this analysis chart


### Line Chart - Weight of Rodents based on Gender discovered per month.


```python
# Bar Chart Legends --
bar_plot_rodent_weight = data_all_taxa_month_gender_weight.plot(figsize=(12,5))

bar_plot_rodent_weight.set_xlabel("Month", fontsize=15)
bar_plot_rodent_weight.set_ylabel("Weight of Rodents", fontsize=15)
bar_plot_rodent_weight.set_title("Weight of Rodents Discovered based on Gender per month.", fontsize=18)
```

##### Graph 1.3 : Graph shows the Weight of Rodents discovered per month over the years, based on Gender. This graph shows an increase of  weight from April till May, and a gradual decrease till July. Then again there is an increase in weight before the beginning of Winter season
